```python
import cv2
import pandas as pd
from PIL import Image

labels = pd.read_csv('/home/sisaha/dsc180b_f1tenth/src/data/annotations/train.csv')

def draw_boxes(image_name):
    selected_value = labels[labels.filename == image_name]
    img = cv2.imread('/home/sisaha/dsc180b_f1tenth/src/data/images/train/{}'.format(image_name))
    for index, row in selected_value.iterrows():
        img = cv2.rectangle(img, (row['xmin'], row['ymin']), (row['xmax'], row['ymax']), (0, 255, 0), 3)
    return img
```

```python
Image.fromarray(draw_boxes('frame0030.jpg'))
```
